# Import Video File 

```python
import cv2

cap = cv2.VideoCapture(r"G:\FB AI & ROBOTICS VIDEOS\10000000_569585997931118_1252944998955129199_n.mp4")
ret, frame = cap.read()
while(1):
   ret, frame = cap.read()
   cv2.imshow('frame',frame)
   if cv2.waitKey(2) & 0xFF == ord('q') or ret==False :
       cap.release()
       cv2.destroyAllWindows()
       break
   cv2.imshow('frame',frame)
```
